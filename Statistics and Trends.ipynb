```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the provided dataset
file_path = 'Airline Dataset.csv'
airline_data = pd.read_csv(file_path)
missing_values = airline_data.isnull().sum()
cleaned_airline_data = airline_data.drop(columns=['Passenger ID', 'First Name', 'Last Name', 'Pilot Name'])
cleaned_airline_data['Departure Date'] = pd.to_datetime(cleaned_airline_data['Departure Date'], errors='coerce')
missing_values_after_cleaning = cleaned_airline_data.isnull().sum()
missing_values, missing_values_after_cleaning, cleaned_airline_data.head()
airline_data.head()

```

```python
# bar chart
plt.figure(figsize=(12, 8))
flight_status_counts = cleaned_airline_data['Flight Status'].value_counts()
sns.barplot(x=flight_status_counts.index, y=flight_status_counts.values)
plt.title('Distribution of Flight Status', fontsize=16)
plt.xlabel('Flight Status', fontsize=14)
plt.ylabel('Number of Flights', fontsize=14)
for index, value in enumerate(flight_status_counts.values):
    plt.text(index, value + 100, f'{value}', ha='center', fontsize=12)

plt.show()
```

```python
# scatter plot

plt.figure(figsize=(12, 8))
sns.scatterplot(data=cleaned_airline_data, x='Age', y='Flight Status', hue='Flight Status', alpha=0.6, s=80)
plt.title('Scatter Plot of Age vs Flight Status', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Flight Status', fontsize=14)
plt.legend(title='Flight Status')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(airline_data['Age'], bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot
plt.figure(figsize=(12, 8))
sns.boxplot(data=cleaned_airline_data, x='Flight Status', y='Age')

# Adding titles and labels for clarity
plt.title('Box Plot of Age by Flight Status', fontsize=16)
plt.xlabel('Flight Status', fontsize=14)
plt.ylabel('Age', fontsize=14)

plt.show()
```
